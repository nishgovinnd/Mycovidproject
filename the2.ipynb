```python
from tkinter import *
from tkinter import ttk
from csv import reader
import csv

with open("Pandamic_Analysis_THE1.csv",'r') as file:
    csv_reader=csv.reader(file)
    d1=list(file)

    
 # click function definition
def click():
    entered_text=textentry.get()
    output.delete(0.0, END)
    try:
        definition = mydictionary[entered_text]
    except:
        definition = "enter a valid state"
    output.insert(END, definition)
# close window function definition    
def close_window():
    window.destroy()
    
window = Tk()
window.title("covid info")
window.configure(background="orange")

photo1=PhotoImage(file='cimg2.gif')
Label(window,image=photo1,bg='black').grid(row=0,column=0,sticky=W)


Label(window, text="covid IFR & CMR calculator for indian states",bg="grey",fg="orange",font="none 15 bold italic").grid(row=2,column=0,sticky=W)

Label(window,text="\n enter the state name",bg="grey",fg="black",font="none 10 italic").grid(row=3,column=0,sticky=W)

Button(window, text='close application',width=20,command=close_window).grid(row=10,column=0,sticky=W)    



states=ttk.Combobox(window)
states['values']=('Maharashtra', 'Tamil Nadu', 'Delhi', 'Gujarat', 'Uttar Pradesh', 'West Bengal', 'Rajasthan', 'Telangana', 'Karnataka', 'Andhra Pradesh', 'Haryana', 'Madhya Pradesh', 'Bihar', 'Assam', 'Jammu Kashmir', 'Odisha', 'Punjab', 'Kerela', 'Uttarakhand', 'Chattisgadh', 'Jharkhand', 'Tripura', 'Goa', 'Manipur', 'Ladakh', 'Himachal pradesh', 'Puducherry', 'Nagaland', 'Chandigadh', 'Dadra and Nagar Haveli', 'Arunachal Pradesh', 'Mizoram', 'Sikkim', 'Andaman and Nicobar', 'Meghalaya', 'Lakshadweep')
states.current(0)
states.grid(column=0,row=5)

infected=tk.Label(window,text="total infected : ",fg="white",bg="dark cyan",font="none 14 bold italic" ).grid(column=2,row=3)
infect=tk.Entry(window,width=24).grid(column=2,row=4)

recovered=tk.Label(window,text="Recovered :",bg="dark cyan",fg="yellow",font="none 14 bold italic").grid(column=2,row=5)
recover=tk.Entry(window,width=24).grid(column=2,row=6)

deaths=tk.Label(window,text="Death : ",bg="dark cyan",fg="red",font="none 14 bold italic").grid(column=2,row=7)
death=tk.Entry(window,width=24).grid(column=2,row=8)

population=tk.Label(window,text='Population:',bg="dark cyan",fg="purple",font="none 14 bold italic").grid(column=2,row=9)
pop=tk.Entry(window,width=24).grid(column=2,row=10)
def cmr():
    x=int(death.get())
    y=int(infect.get())
    return((x/y)*100)
    
def ifr():
    a=int(recover.get())
    b=int(infect.get())
    return((a/b)*100)

    
def cmr_clicked():
    tk.messagebox.showinfo(title="CMR",message=cmr())
    
def ifr_clicked():
    tk.messagebox.showinfo(title="IFR",message=ifr())
    
CMR=tk.Button(window,text="CMR",bg="blue",fg="white",width=10,command=cmr_clicked).grid(column=2,row=11)

IFR=tk.Button(window,text="IFR",bg="yellow",fg="black",width=10,command=ifr_clicked).grid(column=3,row=11)

Label(window,text="CMR result:",fg="blue",font="none 10 italic").grid(row=12,column=2,sticky=W)

Label(window,text="IFR result:",fg="blue",font="none 10 italic").grid(row=14,column=2,sticky=W)

output=Text(window,width=20,height=1,wrap=WORD,background="white")
output.grid(row=13,column=2,sticky=W)

output1=Text(window,width=20,height=1,wrap=WORD,background="white")
output1.grid(row=15,column=2,sticky=W)

window.mainloop()
```
