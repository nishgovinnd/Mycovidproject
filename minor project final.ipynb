```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from csv import reader
import csv
import pandas as pd
import matplotlib.pyplot as plt

data2=pd.read_csv('neigh2.csv',delimiter=',',nrows=150).fillna(0)
data2.plot.area(figsize=(10,5));

plt.title('covid analisys of neighbouring states',fontdict={'fontweight':'bold','fontsize':'30'})
plt.xlabel('days of june',fontdict={'fontweight':'bold','fontsize':'20'})
plt.ylabel('total number of persons',fontdict={'fontweight':'bold','fontsize':'20'})
plt.savefig('covid analisys 3.png',dpi=300)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager
from IPython.core.display import HTML






plt.figure(figsize=(16,8))
s=pd.read_csv('maha2.csv')
plt.plot(s.Date,s.Cured,'b.-',label='cured')
plt.plot(s.Date,s.Deaths,'r.-',label='death')
plt.plot(s.Date,s['Confirmed'],'c.-',label='confirmed')

plt.legend(loc='best')
plt.xticks(s.Date[::5])
plt.title('covid analisys of maharashtra',fontdict={'fontweight':'bold','fontsize':'30'})
plt.xlabel('days of june',fontdict={'fontweight':'bold','fontsize':'20'})
plt.ylabel('number of persons',fontdict={'fontweight':'bold','fontsize':'20'})

plt.savefig('covid analisys5.png',dpi=300)
```
